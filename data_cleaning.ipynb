```python
def classify_time(time_str):
    # Convert string to datetime object
    time = pd.to_datetime(time_str, format='%H:%M').time()
    
    # Define the time categories
    if time < pd.to_datetime('06:00', format='%H:%M').time():
        return 'Early_Morning'
    elif time < pd.to_datetime('12:00', format='%H:%M').time():
        return 'Morning'
    elif time < pd.to_datetime('17:00', format='%H:%M').time():
        return 'Afternoon'
    elif time < pd.to_datetime('21:00', format='%H:%M').time():
        return 'Evening'
    elif time < pd.to_datetime('23:00', format='%H:%M').time():
        return 'Night'
    else:
        return 'Late_Night'
```

```python
# Apply the function to create a new column
data['departure_time'] = data['dep_time'].apply(classify_time)
```

```python
# Define the function to classify the number of stops
def classify_stops(stops_info):
    if 'non-stop' in stops_info.lower():
        return 'zero'
    elif '1-stop' in stops_info.lower():
        return 'one'
    else:
        return 'two_or_more'
    
```

```python
# Apply the function to the DataFrame
data['stop'] = data['stop'].apply(classify_stops)
```

```python
def convert_time_to_float(time_str):
        # Strip leading/trailing whitespace
        time_str = time_str.strip()
        
        # Initialize hours and minutes
        hours = 0
        minutes = 0
        
        # Split the time string based on 'h' and 'm'
        if 'h' in time_str:
            parts = time_str.split('h')
            if '.' in parts[0]:
                parts = parts[0].split('.')
                hours = int(parts[0].strip())
                minutes = int(parts[1].strip())
            else:
                hours = int(parts[0].strip())
                minutes = int(parts[1].split('m')[0].strip())
        elif 'm' in time_str:
            minutes = int(time_str.split('m')[0].strip())
        
        # Convert minutes to a fractional part of an hour
        fractional_minutes = minutes / 60
        
        # Combine hours and fractional minutes
        total = hours + fractional_minutes
        
        return total
```

```python
data['time_taken'] = data['time_taken'].apply(convert_time_to_float)
```

```python
data['time_taken'] = data['time_taken'].round(2)
```

```python
def convert_price_to_int(string_value):
    try:
        # Remove commas and convert to integer
        int_value = int(string_value.replace(',', ''))
        return int_value
    except ValueError as e:
        print(f"Error converting '{string_value}' to integer: {e}")
        return None
```

```python
data['price'] = data['price'].apply(convert_price_to_int)
```
